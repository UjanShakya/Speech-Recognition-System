```python
import librosa
import os
import json
```

```python
DATASET_PATH = "augmented_dataset"
JSON_PATH = "data.json"
SAMPLES_RATES = 22050  # 1 sec worth of sound
```

```python
def prepare_dataset(dataset_path, json_path, num_mfcc = 13, n_fft=2048, hop_length=512):
    # data dictionary
    data = {
        "mappings":[],
        "labels":[],
        "MFCCs":[],
        "files":[]
    }
    
    #loop through all the sub-dirs
    for i , (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):
        
        # we need to ensure thtat we are not at the level
        if dirpath is not dataset_path:
            
            #update mappings
            category = dirpath.split("/")[-1]  #dataset/down -> [dataset,down]
            data["mappings"].append(category)
            print(f"Proccessing {category}")
            
            #loop through all the filenames and extract MFCCs
            for j in filenames:
                
                # get file path
                file_path = os.path.join(dirpath, j)
                
                # load audio file
                signal, sr = librosa.load(file_path)
                
                # ensure the audio file is atleast 1 sec
                if len(signal) >= SAMPLES_RATES:
                    
                    # ensure 1 sec long signal
                    signal = signal[:SAMPLES_RATES]
                    
                    #extract MFCCs
                    
                    MFCCs = librosa.feature.mfcc(signal, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
                    
                    #Store data
                    
                    data["labels"].append(i-1)
                    data["MFCCs"].append(MFCCs.T.tolist())
                    data["files"].append(file_path)
                    print(f"{file_path} : {i-1}")
                  
        
         # store in json file
        with open(json_path, "w") as fp:
                  json.dump(data, fp, indent=4)
            
  
```

```python
if __name__ == "__main__":
    prepare_dataset(DATASET_PATH, JSON_PATH)  
```
